```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
x = np.arange(1, 11)
y = 2*x + 5
plt.plot(x,y)
plt.title('Matplotlib Demo')
plt.xlabel('value of x')
plt.ylabel('value of y')
plt.show()

plt.plot(x,y, 'oc')
plt.show()
```

```python
#Sine Wave Plot
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)
plt.title('Sine Wave')
plt.plot(x, y, 'y')
```

```python
#Plotting different graphs in the same figure using subplot()
x = np.arange(0, 3*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(2, 1, 1)  #(height, width, plot no.)
plt.plot(x, y1)
plt.title('Sine hai yeh!')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Cos hai yeh!')
```

```python
#Bar Graphs
x1 = np.array([-3, -1, 1, 3])
y1 = x1*x1

x2 = np.array([-4, -2, 0, 2, 4])
y2 = x2*x2

plt.bar(x1, y1, align = 'center')
plt.bar(x2, y2, align = 'center', color = 'green')
plt.show()
```

```python
#Histogram
A = [2]
for i in range(3, 100):
    sum = 0
    for j in range(2, i):
        if i%j == 0:
            sum += 1
    if sum == 0:
        A.append(i)
print(A)
A = np.array(A)
print(A)

hist, bins = np.histogram(A, bins = [0, 20, 40, 60, 80, 100])  #ek number numpy function
print(hist)
print(bins)

plt.title('Histogram of Distribution of Prime Numbers from 1 to 100')
plt.hist(A, bins = [0, 19.8, 20.2, 39.8, 40.2, 59.8, 60.2, 79.8, 80.2, 100])
plt.show()
```
